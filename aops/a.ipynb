```python
import random

def abcde(c):
    p = 0.0454545455


    money = 0
    results = []
    for i in range(100000):
        x = 0
        while (random.random() >= p): x += 1
        if c < x: money += c * 0.1
        results.append(x)

    # print(sum(results)/len(results))
    return money/len(results)
```

```python
def zyx(c):
    p = 0.05


    money = 0
    results = []
    for i in range(100000):
        x = 0
        while (random.random() >= p): x += 1
        if c < x: money += 2 * c
        else: money -= c
        results.append(x)

    # print(sum(results)/len(results))
    return money/len(results)
```

```python
x = []
y = []
for c in range(30):
    x.append(c)
    y.append(abcde(c))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({"x": x, "y": y})
```

```python
df.plot(x="x")
```
